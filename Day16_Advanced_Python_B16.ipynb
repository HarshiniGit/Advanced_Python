```python
countries_data = pd.read_csv("countries.csv")
countries_data
```

```python
# To compare the population of US and CHINA

country_1 = countries_data[countries_data.country == 'Afghanistan']
country_1
```

```python
country_2 = countries_data[countries_data.country == 'Zimbabwe']
country_2
```

```python
country_1 = countries_data[countries_data.country == 'Afghanistan']
country_2 = countries_data[countries_data.country == 'Zimbabwe']

plt.plot(country_1.year,country_1.population)
plt.show()
```

```python
x = plt.plot(country_1.year,country_1.population)
y = plt.plot(country_2.year,country_2.population)
plt.xlabel('Population')
plt.xlabel('Census_Year')
plt.title("Comparision of Afghanistan and Zimbabwe population")
plt.legend(["Afghanistan","Zimbabwe"])
plt.show(x,y)
```
